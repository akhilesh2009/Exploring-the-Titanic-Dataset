# **Study Titanic passengers to find what factors improved survival odds.**

**Survival by Gender**

```python
survival_by_gender = train.groupby("Sex")["Survived"].sum()
survival_by_gender
```

Number of Females Survived

```python
females_survive = survival_by_gender["female"]
print(f"Number of females survived: {females_survive}")
```

Number of males Survived

```python
males_survives = survival_by_gender["male"]
print(f"Number of males survived: {males_survives}")
```

**Females have higher rate of survival then males**

Total number of Passengers

```python
total_passengers_survive = (males_survives + females_survive)
print(total_passengers_survive)
```

Percent of Males Survived

```python
percent_males_survived = (males_survives / total_passengers_survive) * 100
print(percent_males_survived)
```

Percent of Females Survived

```python
percent_females_survived = (females_survive / total_passengers_survive) * 100
print(percent_females_survived)
```
